# STEP -2 MFCC FEATURE EXTRACTION

```python
!pip install python_speech_features
```

```python
import soundfile as sf
```

```python
from python_speech_features import mfcc
```

```python
def Mfcc(audiofile):
    s,r=sf.read(audiofile)
    #print(len(s.shape))
    if len(s.shape)==2:
        s=s[0,:]
    #print(len(s.shape))
    x=np.mean(mfcc(s,r, numcep=12,nfft=2048),axis=0)
    return x
```

```python
lengthofmfcc=13
dim=(len(baby_cry+non_babycry),lengthofmfcc)
mydata=np.zeros(dim)
```

```python
for i in range(0,len(baby_cry)):
    data=[]
    mfc=Mfcc(baby_cry[i])
    #print(baby_cry[i]
    #print(mfc)
    for j in mfc:
        data.append(j)
    data.append(1)
    mydata[i,:]=data
all=baby_cry+non_babycry
for i in range(len(baby_cry)-1,len(all)-1):
    data=[]
    mfc=Mfcc(all[i])
    print(all[i])
    for j in mfc:
        data.append(j)
    data.append(0)
    mydata[i,:]=data
```

# STEP -3 DATA SPLIT

```python
input = mydata[:,0:12]
output=mydata[:,12]
```

# STEP -4 DATA SPLIT

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input, output, test_size=0.30, random_state=None)
```

# STEP -5 MODEL BUILDING

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten, Dropout
```

```python
model=Sequential()
model.add(Dense(64, input_dim= 12, init='uniform', activation ='relu'))
model.add(Dense(36,init='uniform', activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

# STEP -6 MODEL TRAINING

```python
model.fit(input,output,validation_data=(X_test,y_test), epochs=10, batch_size=12, verbose=2)
```

# STEP -7 MODEL PREDICTION

```python
y_pred = model.predict(X_test)
```

# STEP -8 MODEL EVALUATION

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred.round()))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred.round()))
```
